```python
x = df[['danceability','energy','loudness','tempo','valence','speechiness','liveness','instrumentalness']]
```

```python
x.isna().sum()
```

```python
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(x)
```

```python
x.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for Spotify Clustering")
plt.show()
```

```python
nc=[1,2,3,4,5,6,7,8,9]
inner=[]
for i in nc:
    km=KMeans(i)
    km.fit(scaled)
    inner.append(km.inertia_)
```

```python
plt.plot(nc,inner,color="black")
plt.scatter(nc,inner,color="red")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(scaled)
```

```python
df['Cluster'] = clusters
```

```python
df.head()
```

```python
df.groupby('Cluster')[['danceability','energy','loudness','tempo','valence','speechiness','liveness','instrumentalness']].mean()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(scaled)

df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(
    x='PCA1',
    y='PCA2',
    hue='Cluster',
    data=df,
    palette='Set2'
)
plt.title("Spotify Song Clustering")
plt.show()
```
